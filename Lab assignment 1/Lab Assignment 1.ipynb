```python
"""
CISB5123 Text Analytics
Lab Assignment 1 - Web Scraping

Student Name: ALSAKKAF MUSAAB
Student ID: SW01083207

Student Name: BAWAZIR AHMED
Student ID: SW01084040

Platform: Trustpilot 
Pages scraped: 5
Output file: trustpilot_lenovo_reviews.csv
"""

import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

base_url = "https://www.trustpilot.com/review/www.lenovo.com?page={}"

headers = {
    "User-Agent": "Mozilla/5.0",
    "Accept-Language": "en-US,en;q=0.9"
}

reviews = []

def scrape_page(page_number):
    """
    This function downloads one Trustpilot review page
    and extracts:
    - reviewer name
    - review date
    - review content
    """
    url = base_url.format(page_number)
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, "html.parser")

    review_cards = soup.find_all("article")

    for card in review_cards:
        name_tag = card.find("span", attrs={"data-consumer-name-typography": "true"})
        date_tag = card.find("time")
        content_tag = card.find("p")

        if name_tag and date_tag and content_tag:
            reviewer_name = name_tag.get_text(strip=True)
            review_date = date_tag.get("datetime", "")[:10]
            review_content = content_tag.get_text(strip=True)

            reviews.append({
                "reviewer_name": reviewer_name,
                "review_date": review_date,
                "review_content": review_content
            })

# Scrape 5 pages
for page in range(1, 6):
    print(f"Scraping page {page}...")
    scrape_page(page)
    time.sleep(1)

# Save to CSV
df = pd.DataFrame(reviews)
df.to_csv("trustpilot_lenovo_reviews.csv", index=False, encoding="utf-8-sig")

print("Scraping completed.")
print("Total reviews collected:", len(df))
```
