```python
import requests
from bs4 import BeautifulSoup

# Download webpage
url = "https://dataquestio.github.io/web-scraping-pages/simple.html"
page = requests.get(url)

# Show status code
print("Status Code:", page.status_code)

# Parse HTML
soup = BeautifulSoup(page.content, 'html.parser')

# Extract paragraph text
content = soup.find('p').get_text()

print("Extracted Content:", content)
```

```python
# Save extracted content into a file
with open("extracted_data1.txt", "w") as file:
    file.write(content)

print("Data saved successfully!")
```

```python
import requests
from bs4 import BeautifulSoup

base_url = "https://quotes.toscrape.com"
url = base_url

einstein_quotes = []

while url:
    page = requests.get(url)
    soup = BeautifulSoup(page.text, 'html.parser')
    
    quotes = soup.find_all('div', class_='quote')
    
    for quote in quotes:
        text = quote.find('span', class_='text').text
        author = quote.find('small', class_='author').text
        
        if author == "Albert Einstein":
            print("Quote:", text)
            print("Author:", author)
            print()
            
            einstein_quotes.append({
                "text": text,
                "author": author
            })
    
    next_button = soup.find('li', class_='next')
    if next_button:
        url = base_url + next_button.find('a')['href']
    else:
        url = None
```

```python
import csv

with open("quotes_Einstein.csv", "w", newline='', encoding='utf-8') as file:
    writer = csv.DictWriter(file, fieldnames=["text", "author"])
    writer.writeheader()
    writer.writerows(einstein_quotes)

print("quotes_Einstein.csv saved successfully!")
```

```python
import requests
from bs4 import BeautifulSoup

base_url = "https://quotes.toscrape.com"
url = base_url

life_quotes = []

while url:
    page = requests.get(url)
    soup = BeautifulSoup(page.text, 'html.parser')
    
    for quote in soup.find_all('div', class_='quote'):
        text = quote.find('span', class_='text').text
        author = quote.find('small', class_='author').text
        
        if "life" in text.lower():
            print("Quote:", text)
            print("Author:", author)
            print()
            
            life_quotes.append({
                "text": text,
                "author": author
            })
    
    next_button = soup.find('li', class_='next')
    url = base_url + next_button.find('a')['href'] if next_button else None
```

```python
import csv

with open("life_quotes.csv", "w", newline='', encoding='utf-8') as file:
    writer = csv.DictWriter(file, fieldnames=["text", "author"])
    writer.writeheader()
    writer.writerows(life_quotes)

print("life_quotes.csv saved successfully!")
```
